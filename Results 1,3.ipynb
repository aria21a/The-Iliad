```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
# change defaults to be less ugly
mpl.rc('xtick', labelsize=14, color="#222222") 
mpl.rc('ytick', labelsize=14, color="#222222") 
mpl.rc('font', **{'family':'sans-serif','sans-serif':['Arial']})
mpl.rc('font', size=16)
mpl.rc('xtick.major', size=6, width=1)
mpl.rc('xtick.minor', size=3, width=1)
mpl.rc('ytick.major', size=6, width=1)
mpl.rc('ytick.minor', size=3, width=1)
mpl.rc('axes', linewidth=1, edgecolor="#222222", labelcolor="#222222")
mpl.rc('text', usetex=False, color="#222222")
```

```python
# Import the csv file as a graph
G = nx.read_edgelist('./Iliad Data.csv', 
                        comments='#',
                        create_using=nx.Graph(), 
                        delimiter=',', 
                        nodetype=str,
                        data=[('weight', float), ('label', str)],
                        encoding='utf-8')
```

```python
# Calculate Modularity
print("Modularity: ")
communities = list(nx.community.greedy_modularity_communities(G, weight='weight'))
nx.community.modularity(G, communities)
```

```python
# Compute betweenness centrality
betweenness = nx.betweenness_centrality(G, weight='weight')

# Sort nodes by betweenness centrality in descending order
sorted_betweenness = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)

# Print sorted results
for node, centrality in sorted_betweenness:
    print(f" {node}: {centrality:.4f}")
```

```python
# Compute weighted degree centrality manually
weighted_degree_centrality = {node: sum(data.get('weight', 1) for _, _, data in G.edges(node, data=True)) for node in G.nodes()}

# Normalize by maximum possible weighted degree
max_degree = max(weighted_degree_centrality.values())
weighted_degree_centrality = {node: centrality / max_degree for node, centrality in weighted_degree_centrality.items()}

# Sort weighted degree centrality
sorted_weighted_degree = sorted(weighted_degree_centrality.items(), key=lambda x: x[1], reverse=True)

# Print sorted weighted degree centrality
for node, centrality in sorted_weighted_degree:
    print(f"{node}: {centrality:.4f}")
```

```python
# Compute closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Sort closeness centrality in descending order
sorted_closeness_centrality = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)

# Print sorted closeness centrality
for node, centrality in sorted_closeness_centrality:
    print(f"{node}: {centrality:.4f}")
```

```python
# Print detected communities
print("Detected Communities:")
for i, community in enumerate(communities):
    print(f"Community {i + 1}: {list(community)}")

# Create a dictionary to map nodes to communities
node_to_community = {}
for i, community in enumerate(communities):
    for node in community:
        node_to_community[node] = i+1
```

```python
# Find the size of each community
community_sizes = {i: len(community) for i, community in enumerate(communities)}

# Define threshold for "small" communities
small_community_threshold = 6

# Identify small communities
small_communities = {i+1 for i, size in community_sizes.items() if size <= small_community_threshold}

print(f"Small communities: {small_communities}")
```

```python
# Identify bridges that connect communities
bridges = []
for node, value in betweenness_centrality.items():
    connected_communities = set(
        node_to_community[neighbor] for neighbor in G.neighbors(node) if neighbor in node_to_community
    )
    
    # Check if the node connects at least one small community
    if len(connected_communities.intersection(small_communities)) > 0 and len(connected_communities) > 1:
        bridges.append((node, len(connected_communities), connected_communities, value))
        
# Sort bridges by betweenness centrality for importance
bridges = sorted(bridges, key=lambda x: x[3], reverse=True)

# Print bridges and the number of communities they connect
for node, count, connected_communities, value in bridges:
    print(f"{node} connects {count} communities with centrality: {value:.4f}")
```

```python
# Print the bridge nodes and the communities they connect
print("\nBridge Nodes Connecting Different Communities:")
for node, count, connected_communities, centrality in bridges:
    print(f"{node} connects {count} communities: {connected_communities}")
```

```python
# Find all combinations of bridge pairs
bridge_pairs = []
for u in range(0, len(bridges)):
    i = 1
    while u + i < len(bridges):
        bridge_pairs.append((bridges[u][0], bridges[u+i][0]))
        i+=1
# Find shortest paths between subplot bridge nodes
for u, v in bridge_pairs:
    path = nx.shortest_path(G, source=u, target=v)
    print(f"Shortest path between {u} and {v}: {path}")
```
